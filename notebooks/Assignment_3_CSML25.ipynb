## **3. M·ª•c ti√™u b√†i to√°n**

* Hi·ªÉu v√† √°p d·ª•ng quy tr√¨nh **pipeline machine learning truy·ªÅn th·ªëng tr√™n d·ªØ li·ªáu ·∫£nh**, bao g·ªìm:

  - Hu·∫•n luy·ªán v√† ƒë√°nh gi√° c√°c b·ªô ph√¢n lo·∫°i c·ªï ƒëi·ªÉn tr√™n ƒë·∫∑c tr∆∞ng tr√≠ch xu·∫•t ƒë∆∞·ª£c.  

* Th·ª±c hi·ªán **ph√¢n t√≠ch EDA** cho t·∫≠p Food-11:

* Th·ª±c h√†nh x√¢y d·ª±ng **pipeline tr√≠ch xu·∫•t ƒë·∫∑c tr∆∞ng (feature-based)**:
  - D√πng c√°c m√¥ h√¨nh pretrained CNN ho·∫∑c Transformer l√†m **b·ªô tr√≠ch xu·∫•t ƒë·∫∑c tr∆∞ng c·ªë ƒë·ªãnh**.

* R√®n luy·ªán kh·∫£ nƒÉng **ph√¢n t√≠ch v√† ƒë√°nh gi√° m√¥ h√¨nh** th√¥ng qua:

  - Ph√¢n t√°ch r√µ c√°c ph·∫ßn: EDA ‚Üí Preprocessing ‚Üí Feature Extraction ‚Üí Model Training ‚Üí Evaluation.  
  - Tr√¨nh b√†y b·∫£ng k·∫øt qu·∫£, bi·ªÉu ƒë·ªì, v√† nh·∫≠n x√©t c√≥ chi·ªÅu s√¢u.

## **4. C√°c m√¥ h√¨nh ƒë∆∞·ª£c tri·ªÉn khai**

### **4.1. Pipeline truy·ªÅn th·ªëng (Feature-based)**

S·ª≠ d·ª•ng c√°c m√¥ h√¨nh **pre-trained** ch·ªâ ƒë·ªÉ tr√≠ch xu·∫•t ƒë·∫∑c tr∆∞ng, sau ƒë√≥ hu·∫•n luy·ªán c√°c **b·ªô ph√¢n lo·∫°i c·ªï ƒëi·ªÉn**.

1. **B·ªô tr√≠ch xu·∫•t ƒë·∫∑c tr∆∞ng CNN:**
   - VGG16 (`imagenet`, `include_top=False`, pooling="avg")  
   - ResNet50 (`imagenet`)  
   - EfficientNetB0 (`imagenet`)  

2. **B·ªô tr√≠ch xu·∫•t ƒë·∫∑c tr∆∞ng Transformer:**
   - ViT (`google/vit-base-patch16-224-in21k`)  
   - Swin Transformer (`microsoft/swin-base-patch4-window7-224`)  

3. **C√°c b·ªô ph√¢n lo·∫°i truy·ªÅn th·ªëng:**
   - Logistic Regression .
   - Linear SVM (LinearSVC).  
   - Random Forest Classifier.  

> **Pipeline t·ªïng qu√°t:**  
> ·∫¢nh ‚Üí Ti·ªÅn x·ª≠ l√Ω ‚Üí CNN/Transformer (Feature Extractor) ‚Üí Vector ƒë·∫∑c tr∆∞ng (.npy) ‚Üí B·ªô ph√¢n lo·∫°i (LR/SVM/RF) ‚Üí D·ª± ƒëo√°n.

---

   - **Chi·∫øn l∆∞·ª£c so s√°nh: Chia th√†nh 3 l·ªõp so s√°nh:**
1. So s√°nh gi·ªØa c√°c c√°ch d√πng c√πng 1 backbone
2. So s√°nh gi·ªØa c√°c backbone kh√°c nhau

> **Pipeline t·ªïng qu√°t:**  
> ·∫¢nh ‚Üí Ti·ªÅn x·ª≠ l√Ω / Augmentation ‚Üí CNN/Transformer ‚Üí Softmax ‚Üí D·ª± ƒëo√°n l·ªõp m√≥n ƒÉn.

```python
splits = ["training", "validation", "evaluation"]

print("S·ªë l∆∞·ª£ng ·∫£nh trong t·ª´ng t·∫≠p:\n")

for split in splits:
    split_path = os.path.join(DATASET_PATH, split)
    if not os.path.exists(split_path):
        continue

    total_images = 0
    for cls in os.listdir(split_path):
        class_path = os.path.join(split_path, cls)
        if os.path.isdir(class_path):
            num_imgs = len([
                f for f in os.listdir(class_path)
                if f.lower().endswith(('.jpg', '.jpeg', '.png'))
            ])
            total_images += num_imgs

    print(f"{split:<12}: {total_images} ·∫£nh")
```

```python
# =====================================
# FEATURE-BASED PIPELINE: CNN + Transformer (batch_size=64 c·ªë ƒë·ªãnh)
# =====================================
!pip install transformers timm --quiet

import os, time, torch, shutil
import numpy as np, pandas as pd
from tqdm import tqdm
from PIL import Image
from torchvision.datasets import ImageFolder
from torchvision import transforms
from torch.utils.data import DataLoader
from transformers import AutoImageProcessor, AutoModel
from tensorflow.keras.applications import (
    VGG16, ResNet50, EfficientNetB0
)
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.resnet50 import preprocess_input as resnet_pre
from tensorflow.keras.applications.vgg16 import preprocess_input as vgg_pre
from tensorflow.keras.applications.efficientnet import preprocess_input as eff_pre
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score

# ========= CONFIG =========
import kagglehub, os, shutil

# T·∫£i dataset t·ª´ Kaggle Hub (n·∫øu ch∆∞a c√≥)
DATASET_PATH = kagglehub.dataset_download("trolukovich/food11-image-dataset")

# Sao ch√©p v√†o /content/food11 (Colab y√™u c·∫ßu ƒë∆∞·ªùng d·∫´n n√†y c·ªë ƒë·ªãnh)
TARGET_PATH = "/content/food11"
if not os.path.exists(TARGET_PATH):
    shutil.copytree(DATASET_PATH, TARGET_PATH)
    print("‚úÖ ƒê√£ sao ch√©p dataset v√†o", TARGET_PATH)
else:
    print("‚ö†Ô∏è Th∆∞ m·ª•c /content/food11 ƒë√£ t·ªìn t·∫°i ‚Äî d√πng l·∫°i.")

# Khai b√°o ƒë∆∞·ªùng d·∫´n cho t·ª´ng t·∫≠p d·ªØ li·ªáu
TRAIN_DIR = os.path.join(TARGET_PATH, "training")
VAL_DIR   = os.path.join(TARGET_PATH, "validation")
TEST_DIR  = os.path.join(TARGET_PATH, "evaluation")

BATCH_SIZE = 64
NUM_WORKERS = 2
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

# =========  DANH S√ÅCH M√î H√åNH =========

transformer_models = {
    "vit":  "google/vit-base-patch16-224-in21k",
    "swin": "microsoft/swin-base-patch4-window7-224"
}

results_all = []

# =========================================
# ========= TRANSFORMER FEATURE ============
# =========================================
@torch.inference_mode()
def forward_to_feature(outputs, model_type):
    if model_type == "vit":
        return outputs.last_hidden_state[:, 0, :]
    elif model_type == "swin":
        return outputs.last_hidden_state.mean(dim=1)
    else:
        if hasattr(outputs, "pooler_output") and outputs.pooler_output is not None:
            return outputs.pooler_output
        return outputs.last_hidden_state.mean(dim=1)

def collate_pil(batch):
    imgs, ys = zip(*batch)
    return list(imgs), torch.tensor(ys)

if RUN_TRANSFORMER:
    for model_key, model_name in transformer_models.items():
        print(f"\n===== Transformer Backbone: {model_key} | Batch={BATCH_SIZE} =====")
        processor = AutoImageProcessor.from_pretrained(model_name)
        model = AutoModel.from_pretrained(model_name).to(device).eval()

        to_pil = transforms.Lambda(lambda x: x if isinstance(x, Image.Image) else Image.fromarray(x))
        train_ds = ImageFolder(TRAIN_DIR, transform=to_pil)
        val_ds   = ImageFolder(VAL_DIR,   transform=to_pil)
        test_ds  = ImageFolder(TEST_DIR,  transform=to_pil)
        train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, collate_fn=collate_pil)
        val_loader   = DataLoader(val_ds,   batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, collate_fn=collate_pil)
        test_loader  = DataLoader(test_ds,  batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, collate_fn=collate_pil)

        def extract(loader):
            feats, labels = [], []
            model_type = getattr(model.config, "model_type", "").lower()
            for imgs, y in tqdm(loader, desc=f"{model_key}"):
                with torch.no_grad():   # ho·∫∑c torch.inference_mode()
                    inputs = processor(images=imgs, return_tensors="pt").to(device)
                    outputs = model(**inputs)
                    emb = forward_to_feature(outputs, model_type)
                    emb = emb.detach().cpu().numpy()
                    feats.append(emb)
                    labels.append(y.numpy())
            return np.concatenate(feats), np.concatenate(labels)


        X_train, y_train = extract(train_loader)
        X_val, y_val = extract(val_loader)
        X_test, y_test = extract(test_loader)

        feat_dir = f"./features/{model_key}"
        os.makedirs(feat_dir, exist_ok=True)
        np.save(os.path.join(feat_dir, "X_train.npy"), X_train)
        np.save(os.path.join(feat_dir, "X_val.npy"), X_val)
        np.save(os.path.join(feat_dir, "X_test.npy"), X_test)
        np.save(os.path.join(feat_dir, "y_train.npy"), y_train)
        np.save(os.path.join(feat_dir, "y_val.npy"), y_val)
        np.save(os.path.join(feat_dir, "y_test.npy"), y_test)

        classifiers = {
            "LogReg": Pipeline([("scaler", StandardScaler()), ("clf", LogisticRegression(max_iter=1000, multi_class="multinomial"))]),
            "SVM": Pipeline([("scaler", StandardScaler()), ("clf", LinearSVC(C=0.1, max_iter=1000))]),
            "RF": RandomForestClassifier(n_estimators=100, max_depth=15, random_state=42)
        }
        for clf_name, clf in classifiers.items():
            print(f"\n--- {model_key} + {clf_name} ---")
            t0 = time.time()
            clf.fit(X_train, y_train)
            t_train = time.time() - t0
            yv_pred = clf.predict(X_val)
            ys_pred = clf.predict(X_test)
            val_acc = accuracy_score(y_val, yv_pred)
            test_acc = accuracy_score(y_test, ys_pred)
            val_f1  = f1_score(y_val, yv_pred, average="macro")
            test_f1 = f1_score(y_test, ys_pred, average="macro")
            results_all.append({
                "Backbone": model_key, "Classifier": clf_name,
                "Val_Acc": val_acc, "Test_Acc": test_acc,
                "Val_F1": val_f1, "Test_F1": test_f1,
                "Train_Time(s)": t_train
            })


```

| Backbone ResNet50 (pretrained, frozen) | Ki·ªÉm tra kh·∫£ nƒÉng tr√≠ch xu·∫•t ƒë·∫∑c tr∆∞ng c·ªßa m√¥ h√¨nh CNN ƒë√£ ƒë∆∞·ª£c hu·∫•n luy·ªán s·∫µn tr√™n ImageNet, khi ch·ªâ hu·∫•n luy·ªán ph·∫ßn ph√¢n lo·∫°i (head). |
| Learning rate = 1e-3 | ƒê√°nh gi√° t·ªëc ƒë·ªô h·ªôi t·ª• v√† ƒë·ªô ·ªïn ƒë·ªãnh khi hu·∫•n luy·ªán head m·ªõi. |
| Batch size = 32 | C√¢n b·∫±ng gi·ªØa t·ªëc ƒë·ªô hu·∫•n luy·ªán v√† ƒë·ªô ·ªïn ƒë·ªãnh c·ªßa gradient. |
| Dropout = 0.3, Dense(256) | Gi·∫£m overfitting v√† tƒÉng kh·∫£ nƒÉng t·ªïng qu√°t h√≥a. |
| Augmentation (xoay, d·ªãch, zoom, l·∫≠t) | TƒÉng ƒëa d·∫°ng d·ªØ li·ªáu, gi√∫p m√¥ h√¨nh h·ªçc ƒë∆∞·ª£c t√≠nh bi·∫øn thi√™n c·ªßa ·∫£nh m√≥n ƒÉn. |
| Callback (EarlyStopping, ReduceLROnPlateau) | Theo d√µi qu√° tr√¨nh h·ªçc, t·ª± ƒë·ªông d·ª´ng khi kh√¥ng c·∫£i thi·ªán v√† gi·∫£m LR ƒë·ªÉ tinh ch·ªânh m√¥ h√¨nh. |
| K√≠ch th∆∞·ªõc ·∫£nh = 224√ó224 | ƒê·∫£m b·∫£o ph√π h·ª£p v·ªõi ki·∫øn tr√∫c ResNet50 v√† ƒë·ªß chi ti·∫øt cho ph√¢n lo·∫°i. |

```python
# C·∫§U H√åNH ·∫¢NH & BATCH
img_size   = (224, 224)
batch_size = 32

# GENERATOR, TI·ªÄN X·ª¨ L√ù, ENCODE, DATA AUGMENT
train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
)

val_test_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input
)

train_gen = train_datagen.flow_from_directory(
    train_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode="categorical",
    shuffle=True,
    seed=SEED
)

val_gen = val_test_datagen.flow_from_directory(
    val_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode="categorical",
    shuffle=False,
    seed=SEED
)

test_gen = val_test_datagen.flow_from_directory(
    eval_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode="categorical",
    shuffle=False,
    seed=SEED
)


num_classes   = train_gen.num_classes
class_indices = train_gen.class_indices
print("S·ªë l·ªõp:", num_classes)
print("class_indices:", class_indices)
```

```python
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models, optimizers, callbacks
import tensorflow as tf



# KH·ªûI T·∫†O BACKBONE RESNET50 (PRETRAINED)

base_model = ResNet50(
    weights="imagenet",
    include_top=False,
    input_shape=(img_size[0], img_size[1], 3)
)

# Freeze to√†n b·ªô backbone
base_model.trainable = False

# head
inputs = tf.keras.Input(shape=(img_size[0], img_size[1], 3))
x = base_model(inputs, training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dense(256, activation="relu")(x)
x = layers.Dropout(0.3)(x)
outputs = layers.Dense(num_classes, activation="softmax")(x)

model_d1 = models.Model(inputs, outputs, name="ResNet50_D1_freeze")

model_d1.compile(
    optimizer=optimizers.Adam(learning_rate=1e-3),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

model_d1.summary()

# CALLBACKS
cb_list = [
    callbacks.EarlyStopping(
        monitor="val_loss",
        patience=5,
        restore_best_weights=True,
        verbose=1
    ),
    callbacks.ReduceLROnPlateau(
        monitor="val_loss",
        factor=0.2,
        patience=3,
        verbose=1
    )
]

# Hu·∫•n luy·ªán
history_d1 = model_d1.fit(
    train_gen,
    epochs=10,
    validation_data=val_gen,
    callbacks=cb_list
)

# ƒê√°nh gi√°
test_loss, test_acc = model_d1.evaluate(test_gen, verbose=1)
print(f"[D1-ResNet50] Test loss: {test_loss:.4f} - Test accuracy: {test_acc:.4f}")
```

**Nh·∫≠n x√©t:**

V·ªõi m√¥ h√¨nh D1 (ResNet50 pretrained, backbone ƒë√≥ng bƒÉng ho√†n to√†n, ch·ªâ hu·∫•n luy·ªán ph·∫ßn head), m√¥ h√¨nh ƒë·∫°t kho·∫£ng 89% accuracy tr√™n t·∫≠p test. ƒê·ªô ch√™nh gi·ªØa train accuracy, validation accuracy v√† test accuracy kh√° nh·ªè, cho th·∫•y m√¥ h√¨nh generalize t·ªët, ch∆∞a c√≥ d·∫•u hi·ªáu overfitting nghi√™m tr·ªçng. Trong 3 epoch ƒë·∫ßu, c·∫£ train/validation loss ƒë·ªÅu gi·∫£m m·∫°nh, sau ƒë√≥ validation loss dao ƒë·ªông nh·∫π quanh 0.43‚Äì0.45, trong khi train loss ti·∫øp t·ª•c gi·∫£m, ph√π h·ª£p v·ªõi k·ª≥ v·ªçng khi ch·ªâ fine-tune ph·∫ßn head c·ªßa m·ªôt backbone ƒë√£ ƒë∆∞·ª£c pretrained tr√™n ImageNet.

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# D·ª± ƒëo√°n tr√™n t·∫≠p TEST
y_pred_probs = model_d1.predict(test_gen)
y_pred = np.argmax(y_pred_probs, axis=1)
y_true = test_gen.classes

# In c√°c ch·ªâ s·ªë ƒë√°nh gi√°
print("="*60)
print("üìä K·∫æT QU·∫¢ ƒê√ÅNH GI√Å TR√äN T·∫¨P TEST (D1 - ResNet50 frozen)")
print("="*60)

acc = accuracy_score(y_true, y_pred)
print(f"Accuracy (Test): {acc:.4f}")

# B√°o c√°o chi ti·∫øt
report = classification_report(
    y_true, y_pred,
    target_names=test_gen.class_indices.keys(),
    digits=4
)
print(report)

# Ma tr·∫≠n nh·∫ßm l·∫´n
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=False, cmap="Blues")
plt.title("Confusion Matrix - Test Set (D1)")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```

Trong m√¥ h√¨nh Test 2, nh√≥m ti·∫øp t·ª•c s·ª≠ d·ª•ng backbone ResNet50 pretrained, nh∆∞ng thay v√¨ gi·ªØ nguy√™n nh∆∞ Test 1, nh√≥m m·ªü (unfreeze) m·ªôt ph·∫ßn c√°c l·ªõp cu·ªëi c·ªßa backbone ƒë·ªÉ fine-tune c√πng v·ªõi ph·∫ßn head m·ªõi.

C·∫•u tr√∫c head v·∫´n gi·ªØ nguy√™n: GlobalAveragePooling2D ‚Üí Dense(256, ReLU) ‚Üí Dropout(0.3) ‚Üí Dense(softmax), nh∆∞ng gi·ªù ƒë√¢y m·ªôt s·ªë l·ªõp cu·ªëi c·ªßa ResNet50 c≈©ng ƒë∆∞·ª£c c·∫≠p nh·∫≠t tr·ªçng s·ªë trong qu√° tr√¨nh hu·∫•n luy·ªán.

Learning rate ƒë∆∞·ª£c gi·∫£m xu·ªëng (v√≠ d·ª• 1e-4) so v·ªõi Test 1, nh·∫±m tr√°nh ‚Äúph√° v·ª°‚Äù c√°c tr·ªçng s·ªë pretrained v√† cho ph√©p m√¥ h√¨nh tinh ch·ªânh nh·∫π nh√†ng c√°c ƒë·∫∑c tr∆∞ng s√¢u, gi√∫p ph√π h·ª£p h∆°n v·ªõi d·ªØ li·ªáu Food11.

Nh·ªù ƒë√≥, Test 2 kh√¥ng ch·ªâ d·ª±a tr√™n ƒë·∫∑c tr∆∞ng ImageNet g·ªëc m√† c√≤n th√≠ch nghi l·∫°i m·ªôt ph·∫ßn backbone cho b√†i to√°n ph√¢n lo·∫°i m√≥n ƒÉn, k·ª≥ v·ªçng c·∫£i thi·ªán ƒë·ªô ch√≠nh x√°c so v·ªõi Test 1.

Test 2 ƒë∆∞·ª£c thi·∫øt k·∫ø v·ªõi c√°c m·ª•c ti√™u:

- ƒê√°nh gi√° hi·ªáu qu·∫£ c·ªßa vi·ªác fine-tune backbone so v·ªõi vi·ªác ch·ªâ hu·∫•n luy·ªán head nh∆∞ Test 1.
‚Üí Ki·ªÉm tra xem vi·ªác c·∫≠p nh·∫≠t th√™m m·ªôt ph·∫ßn tr·ªçng s·ªë trong ResNet50 c√≥ gi√∫p m√¥ h√¨nh ph√¢n bi·ªát t·ªët h∆°n gi·ªØa c√°c l·ªõp Food11 hay kh√¥ng.

- So s√°nh tr·ª±c ti·∫øp v·ªõi Test 1 v·ªÅ ƒë·ªô ch√≠nh x√°c v√† F1-score tr√™n t·∫≠p test.
‚Üí N·∫øu Test 2 cho k·∫øt qu·∫£ t·ªët h∆°n ƒë√°ng k·ªÉ, c√≥ th·ªÉ k·∫øt lu·∫≠n r·∫±ng vi·ªác fine-tune backbone l√† c·∫ßn thi·∫øt trong b√†i to√°n n√†y.

- Quan s√°t ·∫£nh h∆∞·ªüng c·ªßa fine-tune l√™n qu√° tr√¨nh h·ªçc .
‚Üí Ki·ªÉm tra xem m√¥ h√¨nh c√≥ d·ªÖ b·ªã overfitting h∆°n kh√¥ng, v√† li·ªáu EarlyStopping + ReduceLROnPlateau c√≥ gi√∫p ·ªïn ƒë·ªãnh qu√° tr√¨nh hu·∫•n luy·ªán khi backbone b·∫Øt ƒë·∫ßu ƒë∆∞·ª£c c·∫≠p nh·∫≠t.

- L√†m c∆° s·ªü ƒë·ªÉ l·ª±a ch·ªçn chi·∫øn l∆∞·ª£c hu·∫•n luy·ªán cu·ªëi c√πng.
‚Üí T·ª´ so s√°nh Test 1 v√† Test 2, nh√≥m c√≥ th·ªÉ quy·∫øt ƒë·ªãnh n√™n d√πng m√¥ h√¨nh ch·ªâ train head (nhanh, nh·∫π) hay m√¥ h√¨nh fine-tune backbone (ch·∫≠m h∆°n nh∆∞ng ti·ªÅm nƒÉng ch√≠nh x√°c cao h∆°n) cho b∆∞·ªõc tri·ªÉn khai th·ª±c t·∫ø.

```python
# Test 2 - Fine-tune m·ªôt ph·∫ßn ResNet50
base_model = tf.keras.applications.ResNet50(
    weights="imagenet",
    include_top=False,
    input_shape=(img_size[0], img_size[1], 3)
)

# ƒê√≥ng bƒÉng tr∆∞·ªõc, r·ªìi m·ªü kh√≥a ph·∫ßn cu·ªëi
base_model.trainable = True
for layer in base_model.layers[:-40]:
    layer.trainable = False

# X√¢y head nh∆∞ D1
inputs = tf.keras.Input(shape=(img_size[0], img_size[1], 3))
x = base_model(inputs, training=True)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dense(256, activation="relu")(x)
x = layers.Dropout(0.3)(x)
outputs = layers.Dense(num_classes, activation="softmax")(x)

model_d2 = models.Model(inputs, outputs, name="ResNet50_D2_finetune")

# Compile v·ªõi LR th·∫•p
model_d2.compile(
    optimizer=optimizers.Adam(learning_rate=1e-4),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

# Hu·∫•n luy·ªán
history_d2 = model_d2.fit(
    train_gen,
    epochs=10,
    validation_data=val_gen,
    callbacks=cb_list
)

# ƒê√°nh gi√°
test_loss, test_acc = model_d2.evaluate(test_gen, verbose=1)
print(f"[D2-ResNet50 fine-tune] Test loss: {test_loss:.4f} - Test accuracy: {test_acc:.4f}")
```

 **Nh·∫≠n x√©t:**

* Vi·ªác m·ªü fine-tune m·ªôt ph·∫ßn ResNet50 cho ph√©p m√¥ h√¨nh h·ªçc l·∫°i nh·ªØng ƒë·∫∑c tr∆∞ng trung-cao c·∫•p (texture, pattern ƒë·∫∑c tr∆∞ng c·ªßa m√≥n ƒÉn), v·ªën backbone g·ªëc ImageNet ch∆∞a t·ªëi ∆∞u cho th·ª±c ph·∫©m.

* C·∫£i thi·ªán r√µ r·ªát ·ªü c√°c l·ªõp ‚Äúg·∫ßn nhau v·ªÅ texture‚Äù nh∆∞ Bread, Dessert, Fried food, Meat ‚Äî nh·ªØng l·ªõp m√† D1 th∆∞·ªùng nh·∫ßm.

* Callback ReduceLROnPlateau ph√°t huy t·ªët: khi val_loss ng·ª´ng gi·∫£m, LR h·∫° xu·ªëng gi√∫p m√¥ h√¨nh tinh ch·ªânh s√¢u m√† kh√¥ng ph√° v·ª° tr·ªçng s·ªë pretrained.

K·∫øt qu·∫£ ch·ª©ng minh r·∫±ng fine-tune c√≥ ki·ªÉm so√°t l√† chi·∫øn l∆∞·ª£c hi·ªáu qu·∫£ ƒë·ªÉ n√¢ng cao hi·ªáu nƒÉng trong pipeline h·ªçc s√¢u end-to-end

## Test 3

 Trong m√¥ h√¨nh Test 3, nh√≥m ti·∫øp t·ª•c s·ª≠ d·ª•ng backbone ResNet50 pretrained tr√™n ImageNet, nh∆∞ng thay v√¨ ch·ªâ m·ªü m·ªôt ph·∫ßn nh∆∞ Test 2, to√†n b·ªô c√°c l·ªõp c·ªßa backbone ƒë∆∞·ª£c unfreeze v√† c·∫≠p nh·∫≠t trong qu√° tr√¨nh hu·∫•n luy·ªán.

 ƒêi·ªÅu n√†y bi·∫øn m√¥ h√¨nh th√†nh m·ªôt pipeline end-to-end th·ª±c s·ª±, n∆°i m·ªçi t·∫ßng ƒë·ªÅu tham gia h·ªçc l·∫°i, cho ph√©p m·∫°ng tinh ch·ªânh s√¢u h∆°n c√°c ƒë·∫∑c tr∆∞ng ph√π h·ª£p v·ªõi d·ªØ li·ªáu m√≥n ƒÉn (Food11).

 C·∫•u tr√∫c head gi·ªØ nguy√™n nh∆∞ c√°c test tr∆∞·ªõc: GlobalAveragePooling2D ‚Üí Dense(256, ReLU) ‚Üí Dropout(0.3) ‚Üí Dense(softmax).

Tuy nhi√™n, v√¨ to√†n b·ªô backbone ƒë∆∞·ª£c m·ªü kh√≥a, nh√≥m gi·∫£m learning rate ƒë√°ng k·ªÉ (kho·∫£ng 1e-5) ƒë·ªÉ tr√°nh ‚Äúl√†m h·ªèng‚Äù c√°c tr·ªçng s·ªë pretrained, ƒë·ªìng th·ªùi s·ª≠ d·ª•ng EarlyStopping v√† ReduceLROnPlateau nh·∫±m ki·ªÉm so√°t qu√° tr√¨nh h·ªôi t·ª• v√† h·∫°n ch·∫ø overfitting.

 M√¥ h√¨nh ƒë∆∞·ª£c hu·∫•n luy·ªán c√πng pipeline d·ªØ li·ªáu, augmentation, batch size v√† preprocessing gi·ªëng c√°c test tr∆∞·ªõc ƒë·ªÉ ƒë·∫£m b·∫£o t√≠nh so s√°nh c√¥ng b·∫±ng.

**Test 3 ƒë∆∞·ª£c thi·∫øt k·∫ø ƒë·ªÉ:**

- ƒê√°nh gi√° m·ª©c ·∫£nh h∆∞·ªüng c·ªßa fine-tune to√†n ph·∫ßn: Ki·ªÉm tra xem vi·ªác c·∫≠p nh·∫≠t to√†n b·ªô tr·ªçng s·ªë c√≥ gi√∫p m√¥ h√¨nh h·ªçc ƒë∆∞·ª£c ƒë·∫∑c tr∆∞ng th·ªã gi√°c chuy√™n bi·ªát c·ªßa Food11 t·ªët h∆°n so v·ªõi fine-tune m·ªôt ph·∫ßn (Test 2).
- Quan s√°t hi·ªáu nƒÉng ‚Äì ƒë·ªô ·ªïn ƒë·ªãnh: So s√°nh accuracy, recall, F1-score v√† loss tr√™n t·∫≠p test v·ªõi hai m√¥ h√¨nh tr∆∞·ªõc; ƒë·ªìng th·ªùi theo d√µi xem vi·ªác m·ªü to√†n b·ªô backbone c√≥ g√¢y dao ƒë·ªông ho·∫∑c overfitting.
- Ph√¢n t√≠ch chi ph√≠ v√† hi·ªáu qu·∫£: X√°c ƒë·ªãnh m·ª©c ƒë√°nh ƒë·ªïi gi·ªØa hi·ªáu nƒÉng ƒë·∫°t ƒë∆∞·ª£c v√† th·ªùi gian hu·∫•n luy·ªán tƒÉng th√™m do s·ªë l∆∞·ª£ng tham s·ªë l·ªõn h∆°n nhi·ªÅu.
- L√†m c∆° s·ªü ch·ªçn chi·∫øn l∆∞·ª£c t·ªëi ∆∞u: K·∫øt h·ª£p k·∫øt qu·∫£ ba m√¥ h√¨nh (D1 ‚Äì frozen, D2 ‚Äì partial fine-tune, D3 ‚Äì full fine-tune) ƒë·ªÉ x√°c ƒë·ªãnh c·∫•u h√¨nh ph√π h·ª£p nh·∫•t cho pipeline end-to-end, c√¢n b·∫±ng gi·ªØa ƒë·ªô ch√≠nh x√°c, t·ªëc ƒë·ªô v√† kh·∫£ nƒÉng t·ªïng qu√°t h√≥a.

---

| N·ªôi dung th·ª≠ nghi·ªám | M·ª•c ti√™u c·ª• th·ªÉ |
|---------------------|-----------------|
| Backbone ResNet50 (pretrained, unfreeze to√†n b·ªô) | Ki·ªÉm tra hi·ªáu qu·∫£ c·ªßa vi·ªác c·∫≠p nh·∫≠t to√†n b·ªô tr·ªçng s·ªë backbone trong hu·∫•n luy·ªán end-to-end, ƒë√°nh gi√° kh·∫£ nƒÉng h·ªçc s√¢u ƒë·∫∑c tr∆∞ng ri√™ng c·ªßa Food11. |
| Learning rate = 1e-5 | H·∫° LR th·∫•p ƒë·ªÉ tr√°nh ph√° v·ª° tr·ªçng s·ªë pretrained khi to√†n b·ªô m·∫°ng ƒë∆∞·ª£c c·∫≠p nh·∫≠t. |
| Batch size = 32 | Gi·ªØ nguy√™n ƒë·ªÉ ƒë·∫£m b·∫£o t√≠nh so s√°nh v·ªõi c√°c m√¥ h√¨nh D1‚ÄìD2. |
| Dropout = 0.3, Dense(256) | Ti·∫øp t·ª•c d√πng regularization nh·∫±m gi·∫£m overfitting khi s·ªë l∆∞·ª£ng tham s·ªë hu·∫•n luy·ªán tƒÉng ƒë√°ng k·ªÉ. |
| Augmentation (xoay, d·ªãch, zoom, l·∫≠t) | Cung c·∫•p th√™m bi·∫øn th·ªÉ h√¨nh ·∫£nh ƒë·ªÉ h·ªó tr·ª£ m√¥ h√¨nh h·ªçc t·ªïng qu√°t h√≥a t·ªët h∆°n trong fine-tune to√†n ph·∫ßn. |
| Callback (EarlyStopping, ReduceLROnPlateau) | Ki·ªÉm so√°t qu√° tr√¨nh h·ªôi t·ª•, gi√∫p m√¥ h√¨nh d·ª´ng ƒë√∫ng l√∫c khi to√†n b·ªô backbone b·∫Øt ƒë·∫ßu ƒëi·ªÅu ch·ªânh tr·ªçng s·ªë. |
| K√≠ch th∆∞·ªõc ·∫£nh = 224√ó224 | Gi·ªØ nguy√™n ƒë·ªÉ duy tr√¨ ƒë·ªô chi ti·∫øt v√† ƒë·∫£m b·∫£o t√≠nh nh·∫•t qu√°n trong pipeline hu·∫•n luy·ªán. |

```python
# Test 3 - Fine-tune to√†n b·ªô
SEED = 42
os.environ["PYTHONHASHSEED"] = str(SEED)
os.environ["TF_DETERMINISTIC_OPS"] = "1"
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

base_model = ResNet50(
    weights="imagenet",
    include_top=False,
    input_shape=(img_size[0], img_size[1], 3)
)
base_model.trainable = True

# X√¢y head nh∆∞ D1
inputs = tf.keras.Input(shape=(img_size[0], img_size[1], 3))
x = base_model(inputs, training=True)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dense(256, activation="relu")(x)
x = layers.Dropout(0.3)(x)
outputs = layers.Dense(num_classes, activation="softmax")(x)

model_d3 = models.Model(inputs, outputs, name="ResNet50_D3_full_finetune")

# Compile v·ªõi LR th·∫•p
model_d3.compile(
    optimizer=optimizers.Adam(learning_rate=1e-5),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

# Hu·∫•n luy·ªán
history_d3 = model_d3.fit(
    train_gen,
    epochs=10,
    validation_data=val_gen,
    callbacks=cb_list
)

# ƒê√°nh gi√°
test_loss_d3, test_acc_d3 = model_d3.evaluate(test_gen, verbose=1)
print(f"[D3-ResNet50 full fine-tune] Test loss: {test_loss_d3:.4f} - Test accuracy: {test_acc_d3:.4f}")
```

- D3 v·∫´n c√≥ ∆∞u ƒëi·ªÉm ·ªü ƒë·ªô ·ªïn ƒë·ªãnh v√† h·ªôi t·ª• m∆∞·ª£t, nh∆∞ng hi·ªáu nƒÉng test ch·ªâ nh·ªânh h∆°n D1, k√©m D2 nh·∫π ‚Üí th·ªÉ hi·ªán r·∫±ng ‚Äúpartial fine-tune‚Äù (D2) l√† ƒëi·ªÉm c√¢n b·∫±ng t·ªëi ∆∞u gi·ªØa t·ªëc ƒë·ªô, ƒë·ªô ch√≠nh x√°c v√† kh·∫£ nƒÉng t·ªïng qu√°t h√≥a.

- Callback ho·∫°t ƒë·ªông hi·ªáu qu·∫£, gi√∫p m√¥ h√¨nh d·ª´ng ƒë√∫ng l√∫c, kh√¥ng overfit d√π s·ªë tham s·ªë tƒÉng g·∫•p ƒë√¥i so v·ªõi D2.

K·∫øt qu·∫£ Test 3 cho th·∫•y vi·ªác fine-tune to√†n b·ªô backbone ResNet50 gi√∫p m√¥ h√¨nh h·ªçc s√¢u h∆°n v√† gi·∫£m loss so v·ªõi m√¥ h√¨nh frozen, nh∆∞ng m·ª©c c·∫£i thi·ªán kh√¥ng v∆∞·ª£t tr·ªôi so v·ªõi fine-tune m·ªôt ph·∫ßn. D3 ƒë·∫°t ƒë·ªô ch√≠nh x√°c cao tr√™n t·∫≠p test, th·ªÉ hi·ªán kh·∫£ nƒÉng th√≠ch nghi t·ªët nh∆∞ng chi ph√≠ hu·∫•n luy·ªán cao h∆°n ƒë√°ng k·ªÉ. Qua ƒë√≥, c√≥ th·ªÉ k·∫øt lu·∫≠n r·∫±ng vi·ªác fine-tune to√†n ph·∫ßn kh√¥ng c·∫ßn thi·∫øt trong b√†i to√°n n√†y; chi·∫øn l∆∞·ª£c m·ªü m·ªôt ph·∫ßn backbone (nh∆∞ D2) ƒë√£ ƒë·ªß t·ªëi ∆∞u cho b√†i to√°n ph√¢n lo·∫°i Food11

```python
# ƒê√°nh gi√° D3
test_loss_d3, test_acc_d3 = model_d3.evaluate(test_gen, verbose=1)
print(f"[D3-ResNet50 full fine-tune] Test loss: {test_loss_d3:.4f} - Test accuracy: {test_acc_d3:.4f}")

# D·ª± ƒëo√°n tr√™n t·∫≠p TEST b·∫±ng D3
y_pred_probs_d3 = model_d3.predict(test_gen)
y_pred_d3 = np.argmax(y_pred_probs_d3, axis=1)
y_true = test_gen.classes

print("="*60)
print("üìä K·∫æT QU·∫¢ ƒê√ÅNH GI√Å TR√äN T·∫¨P TEST (D3 - ResNet50 full fine-tuned)")
print("="*60)

acc_d3 = accuracy_score(y_true, y_pred_d3)
print(f"Accuracy (Test): {acc_d3:.4f}")

report_d3 = classification_report(
    y_true,
    y_pred_d3,
    target_names=test_gen.class_indices.keys(),
    digits=4
)
print(report_d3)

cm_d3 = confusion_matrix(y_true, y_pred_d3)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_d3, annot=False, cmap="Greens")
plt.title("Confusion Matrix - Test Set (D3 - Full fine-tuned)")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```

- Rice v√† Soup l√† hai l·ªõp ƒë·∫°t ƒë·ªô ph√¢n bi·ªát ho√†n h·∫£o ‚Üí bi·ªÉu hi·ªán c·ªßa feature h·ªçc s√¢u ƒë·∫∑c tr∆∞ng v√πng n·ªÅn v√† h√¨nh th√°i m√≥n ƒÉn.

**K·∫øt lu·∫≠n:**

M√¥ h√¨nh D3 th·ªÉ hi·ªán nƒÉng l·ª±c h·ªçc ƒë·∫∑c tr∆∞ng m·∫°nh v√† ·ªïn ƒë·ªãnh khi fine-tune to√†n b·ªô backbone. Tuy nhi√™n, l·ª£i √≠ch thu ƒë∆∞·ª£c so v·ªõi D2 l√† kh√¥ng ƒë√°ng k·ªÉ trong khi chi ph√≠ hu·∫•n luy·ªán cao h∆°n ƒë√°ng k·ªÉ. Nh∆∞ v·∫≠y, fine-tune m·ªôt ph·∫ßn backbone (D2) v·∫´n l√† chi·∫øn l∆∞·ª£c hi·ªáu qu·∫£ nh·∫•t gi·ªØa ƒë·ªô ch√≠nh x√°c, chi ph√≠ v√† kh·∫£ nƒÉng t·ªïng qu√°t ho√°.

## Test 4

Trong m√¥ h√¨nh Test 4, nh√≥m thay backbone ResNet50 b·∫±ng EfficientNetB0 ‚Äì m·ªôt ki·∫øn tr√∫c CNN hi·ªán ƒë·∫°i, ƒë∆∞·ª£c thi·∫øt k·∫ø t·ªëi ∆∞u gi·ªØa ƒë·ªô ch√≠nh x√°c v√† s·ªë l∆∞·ª£ng tham s·ªë. T∆∞∆°ng t·ª± Test 2, backbone EfficientNetB0 ƒë∆∞·ª£c kh·ªüi t·∫°o t·ª´ tr·ªçng s·ªë pretrained tr√™n ImageNet, sau ƒë√≥ ch·ªâ unfreeze m·ªôt ph·∫ßn c√°c l·ªõp cu·ªëi ƒë·ªÉ fine-tune c√πng v·ªõi ph·∫ßn head m·ªõi.

C·∫•u tr√∫c head v·∫´n ƒë∆∞·ª£c gi·ªØ nguy√™n nh∆∞ c√°c test tr∆∞·ªõc:
GlobalAveragePooling2D ‚Üí Dense(256, ReLU) ‚Üí Dropout(0.3) ‚Üí Dense(softmax).
Nh·ªù ƒë√≥, m·ªçi kh√°c bi·ªát v·ªÅ hi·ªáu nƒÉng ch·ªß y·∫øu ƒë·∫øn t·ª´ ki·∫øn tr√∫c backbone ch·ª© kh√¥ng do thay ƒë·ªïi ·ªü ph·∫ßn classifier.

ƒê·ªÉ ph√π h·ª£p v·ªõi EfficientNet, pipeline d·ªØ li·ªáu c·ªßa Test 4 s·ª≠ d·ª•ng h√†m ti·ªÅn x·ª≠ l√Ω ri√™ng eff_preprocess (thay cho preprocess c·ªßa ResNet), trong khi c√°c y·∫øu t·ªë kh√°c nh∆∞ k√≠ch th∆∞·ªõc ·∫£nh 224√ó224, batch size, augmentation, s·ªë epoch, callback ƒë·ªÅu ƒë∆∞·ª£c gi·ªØ gi·ªëng Test 2. Learning rate ƒë∆∞·ª£c ƒë·∫∑t ·ªü m·ª©c 1e-4, ƒë·ªß nh·ªè ƒë·ªÉ tinh ch·ªânh m·ªôt ph·∫ßn backbone m√† kh√¥ng l√†m h·ªèng tr·ªçng s·ªë pretrained.

---
Test 4 ƒë∆∞·ª£c thi·∫øt k·∫ø v·ªõi m·ª•c ti√™u:

- So s√°nh ki·∫øn tr√∫c backbone: ƒê√°nh gi√° xem EfficientNetB0 ‚Äì v·ªõi s·ªë tham s·ªë √≠t h∆°n nh∆∞ng ƒë∆∞·ª£c thi·∫øt k·∫ø t·ªëi ∆∞u ‚Äì c√≥ th·ªÉ ƒë·∫°t hi·ªáu nƒÉng t∆∞∆°ng ƒë∆∞∆°ng ho·∫∑c t·ªët h∆°n ResNet50 (Test 2) khi s·ª≠ d·ª•ng c√πng chi·∫øn l∆∞·ª£c fine-tune m·ªôt ph·∫ßn.

- ƒêo l∆∞·ªùng hi·ªáu qu·∫£‚Äìchi ph√≠: Xem x√©t trade-off gi·ªØa ƒë·ªô ch√≠nh x√°c, t·ªëc ƒë·ªô h·ªôi t·ª• v√† th·ªùi gian hu·∫•n luy·ªán tr√™n m·ªói epoch, t·ª´ ƒë√≥ ƒë√°nh gi√° t√≠nh ph√π h·ª£p c·ªßa EfficientNetB0 cho b√†i to√°n ph√¢n lo·∫°i m√≥n ƒÉn.

- Ki·ªÉm tra kh·∫£ nƒÉng t·ªïng qu√°t ho√°: So s√°nh accuracy, F1-score v√† confusion matrix v·ªõi c√°c test tr∆∞·ªõc, ƒë·∫∑c bi·ªát ·ªü c√°c l·ªõp kh√≥ nh∆∞ Bread, Dairy product, Fried food v√† Seafood.

- L√†m c∆° s·ªü ch·ªçn backbone cu·ªëi c√πng: K·∫øt h·ª£p k·∫øt qu·∫£ Test 2 (ResNet50 partial fine-tune) v√† Test 4 (EfficientNetB0 partial fine-tune) ƒë·ªÉ x√°c ƒë·ªãnh backbone n√†o ph√π h·ª£p h∆°n cho pipeline end-to-end v·ªÅ c·∫£ hi·ªáu nƒÉng l·∫´n t√†i nguy√™n.

---
| N·ªôi dung th·ª≠ nghi·ªám | M·ª•c ti√™u c·ª• th·ªÉ |
|---------------------|-----------------|
| Backbone EfficientNetB0 (pretrained, unfreeze m·ªôt ph·∫ßn l·ªõp cu·ªëi) | ƒê√°nh gi√° hi·ªáu qu·∫£ c·ªßa ki·∫øn tr√∫c EfficientNetB0 so v·ªõi ResNet50 khi c√πng s·ª≠ d·ª•ng chi·∫øn l∆∞·ª£c fine-tune m·ªôt ph·∫ßn backbone. |
| Learning rate = 1e-4 | Cho ph√©p tinh ch·ªânh nh·∫π nh√†ng c√°c l·ªõp cu·ªëi c·ªßa EfficientNetB0 m√† kh√¥ng l√†m h·ªèng tr·ªçng s·ªë pretrained. |
| Batch size = 32 | Gi·ªØ nguy√™n ƒë·ªÉ ƒë·∫£m b·∫£o t√≠nh so s√°nh c√¥ng b·∫±ng v·ªõi c√°c test d√πng ResNet50. |
| Dropout = 0.3, Dense(256) | H·∫°n ch·∫ø overfitting cho ph·∫ßn head trong b·ªëi c·∫£nh s·ªë tham s·ªë trainable tƒÉng th√™m t·ª´ backbone. |
| Augmentation (xoay, d·ªãch, zoom, l·∫≠t) | TƒÉng ƒëa d·∫°ng d·ªØ li·ªáu, gi√∫p EfficientNetB0 h·ªçc ƒë∆∞·ª£c t√≠nh bi·∫øn thi√™n v·ªÅ g√≥c ch·ª•p, v·ªã tr√≠ v√† scale c·ªßa m√≥n ƒÉn. |
| Ti·ªÅn x·ª≠ l√Ω eff_preprocess | ƒê·∫£m b·∫£o ·∫£nh ƒë·∫ßu v√†o ph√π h·ª£p v·ªõi chu·∫©n m√† EfficientNetB0 ƒë∆∞·ª£c pretrained tr√™n ImageNet, t·∫≠n d·ª•ng t·ªëi ƒëa tr·ªçng s·ªë h·ªçc s·∫µn. |
| Callback (EarlyStopping, ReduceLROnPlateau) | Theo d√µi qu√° tr√¨nh h·ªçc, t·ª± ƒë·ªông d·ª´ng khi kh√¥ng c√≤n c·∫£i thi·ªán v√† gi·∫£m LR khi val_loss ch·ªØng l·∫°i, gi√∫p m√¥ h√¨nh h·ªôi t·ª• ·ªïn ƒë·ªãnh. |
| K√≠ch th∆∞·ªõc ·∫£nh = 224√ó224 | Gi·ªØ nguy√™n nh∆∞ c√°c test tr∆∞·ªõc, ƒë·∫£m b·∫£o ƒë·ªß chi ti·∫øt cho ph√¢n lo·∫°i v√† t∆∞∆°ng th√≠ch v·ªõi ki·∫øn tr√∫c EfficientNetB0. |

```python
from tensorflow.keras.applications.efficientnet import EfficientNetB0, preprocess_input as eff_preprocess

# GENERATOR RI√äNG CHO D4 - EfficientNet
eff_train_datagen = ImageDataGenerator(
    preprocessing_function=eff_preprocess,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
)

eff_val_test_datagen = ImageDataGenerator(
    preprocessing_function=eff_preprocess
)

eff_train_gen = eff_train_datagen.flow_from_directory(
    train_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode="categorical",
    shuffle=True,
    seed=SEED
)

eff_val_gen = eff_val_test_datagen.flow_from_directory(
    val_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode="categorical",
    shuffle=False,
    seed=SEED
)

eff_test_gen = eff_val_test_datagen.flow_from_directory(
    eval_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode="categorical",
    shuffle=False,
    seed=SEED
)

num_classes   = eff_train_gen.num_classes
class_indices = eff_train_gen.class_indices
print("S·ªë l·ªõp:", num_classes)
print("class_indices:", class_indices)

# Kh·ªüi t·∫°o backbone EfficientNetB0 pretrained tr√™n ImageNet
base_model = EfficientNetB0(
    weights="imagenet",
    include_top=False,
    input_shape=(img_size[0], img_size[1], 3)
)

# Cho ph√©p train, nh∆∞ng ch·ªâ m·ªü m·ªôt ph·∫ßn c√°c layer cu·ªëi
base_model.trainable = True
for layer in base_model.layers[:-40]:   # freeze t·∫•t c·∫£ tr·ª´ ~40 layer cu·ªëi
    layer.trainable = False

# X√¢y head gi·ªëng D1/D2/D3 ƒë·ªÉ so s√°nh c√¥ng b·∫±ng
inputs = tf.keras.Input(shape=(img_size[0], img_size[1], 3))
x = base_model(inputs, training=True)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dense(256, activation="relu")(x)
x = layers.Dropout(0.3)(x)
outputs = layers.Dense(num_classes, activation="softmax")(x)

model_d4 = models.Model(inputs, outputs, name="EffNetB0_D4_partial_finetune")

model_d4.compile(
    optimizer=optimizers.Adam(learning_rate=1e-4),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

model_d4.summary()

cb_list = [
    callbacks.EarlyStopping(
        monitor="val_loss",
        patience=3,
        restore_best_weights=True,
        verbose=1
    ),
    callbacks.ReduceLROnPlateau(
        monitor="val_loss",
        factor=0.2,
        patience=3,
        verbose=1
    )
]

history_d4 = model_d4.fit(
    eff_train_gen,
    epochs=10,
    validation_data=eff_val_gen,
    callbacks=cb_list
)

test_loss_d4, test_acc_d4 = model_d4.evaluate(eff_test_gen, verbose=1)
print(f"[D4-EfficientNetB0 partial fine-tune] Test loss: {test_loss_d4:.4f} - Test accuracy: {test_acc_d4:.4f}")
```

```python


# D·ª± ƒëo√°n tr√™n t·∫≠p TEST b·∫±ng D4
y_pred_probs_d4 = model_d4.predict(eff_test_gen)
y_pred_d4 = np.argmax(y_pred_probs_d4, axis=1)
y_true = eff_test_gen.classes   # do eff_test_gen ƒë√£ shuffle=False

print("="*60)
print("üìä K·∫æT QU·∫¢ ƒê√ÅNH GI√Å TR√äN T·∫¨P TEST (D4 - EfficientNetB0 partial fine-tune)")
print("="*60)

acc_d4 = accuracy_score(y_true, y_pred_d4)
print(f"Accuracy (Test): {acc_d4:.4f}")

# B√°o c√°o chi ti·∫øt t·ª´ng l·ªõp
report_d4 = classification_report(
    y_true,
    y_pred_d4,
    target_names=eff_test_gen.class_indices.keys(),
    digits=4
)
print(report_d4)

# Ma tr·∫≠n nh·∫ßm l·∫´n
cm_d4 = confusion_matrix(y_true, y_pred_d4)

plt.figure(figsize=(8, 6))
sns.heatmap(cm_d4, annot=False, fmt="d", cmap="Greens")
plt.title("Confusion Matrix - Test Set (D4 - EfficientNetB0 partial fine-tune)")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```

```python
from tensorflow.keras.applications.efficientnet import EfficientNetB0, preprocess_input as eff_preprocess

# ====== GENERATOR ======
eff_train_datagen = ImageDataGenerator(
    preprocessing_function=eff_preprocess,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
)

eff_val_test_datagen = ImageDataGenerator(
    preprocessing_function=eff_preprocess
)

eff_train_gen = eff_train_datagen.flow_from_directory(
    train_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode="categorical",
    shuffle=True,
    seed=SEED
)

eff_val_gen = eff_val_test_datagen.flow_from_directory(
    val_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode="categorical",
    shuffle=False,
    seed=SEED
)

eff_test_gen = eff_val_test_datagen.flow_from_directory(
    eval_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode="categorical",
    shuffle=False,
    seed=SEED
)

num_classes   = eff_train_gen.num_classes
class_indices = eff_train_gen.class_indices
print("S·ªë l·ªõp:", num_classes)
print("class_indices:", class_indices)

# ====== BACKBONE ======
base_model = EfficientNetB0(
    weights="imagenet",
    include_top=False,
    input_shape=(img_size[0], img_size[1], 3)
)

# Cho ph√©p train, nh∆∞ng ch·ªâ m·ªü m·ªôt ph·∫ßn (40 l·ªõp cu·ªëi)
base_model.trainable = True
for layer in base_model.layers[:-40]:
    layer.trainable = False

# ====== HEAD (thay ƒë·ªïi) ======
inputs = tf.keras.Input(shape=(img_size[0], img_size[1], 3))
x = base_model(inputs, training=True)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dense(512, activation="relu")(x)      # tƒÉng dung l∆∞·ª£ng head
x = layers.Dropout(0.4)(x)                       # regularization m·∫°nh h∆°n
outputs = layers.Dense(num_classes, activation="softmax")(x)

model_d7 = models.Model(inputs, outputs, name="EffNetB0_D7_dense512_dropout04")

# ====== COMPILE ======
model_d7.compile(
    optimizer=optimizers.Adam(learning_rate=1e-4),  # gi·ªØ nguy√™n LR c·ªßa D4
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

model_d7.summary()

# ====== CALLBACKS ======
cb_list = [
    callbacks.EarlyStopping(
        monitor="val_loss",
        patience=3,
        restore_best_weights=True,
        verbose=1
    ),
    callbacks.ReduceLROnPlateau(
        monitor="val_loss",
        factor=0.2,
        patience=3,
        verbose=1
    )
]

# ====== TRAIN ======
history_d7 = model_d7.fit(
    eff_train_gen,
    epochs=10,
    validation_data=eff_val_gen,
    callbacks=cb_list
)

# ====== EVALUATE ======
test_loss_d7, test_acc_d7 = model_d7.evaluate(eff_test_gen, verbose=1)
print(f"[D7-EfficientNetB0 Dense512 Dropout0.4] Test loss: {test_loss_d7:.4f} - Test accuracy: {test_acc_d7:.4f}")
```
